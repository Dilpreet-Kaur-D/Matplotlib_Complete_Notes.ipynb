## Matplotlib

Matplotlib is an open-source Python library for data visualization. It allows us to create both simple and highly customized plots, including line plots, bar charts, scatter plots, histograms, and pie charts. It is widely used in data analysis, machine learning, and scientific computing to visualize data and communicate insights. Seaborn is built on top of Matplotlib and provides a higher-level interface for creating more attractive statistical visualizations.
<!--  -->
Besides creating graphs, Matplotlib allows extensive customization, supports multiple plots in a single figure, saves visualizations in various formats, creates animations, plots mathematical functions, adds annotations, and produces publication-quality figures.

#### Matplotlib has two styles
<!--  -->
 ##### Pyplot Style: 
 A simple, state-based interface that creates plots using the plt module.
<!--  -->
 ##### Object-Oriented Style:
 A flexible approach that creates and customizes plots using Figure and Axes objects.

| Pyplot Style    | Object-Oriented Style |
| --------------- | --------------------- |
| `plt.plot()`    | `ax.plot()`           |
| `plt.scatter()` | `ax.scatter()`        |
| `plt.bar()`     | `ax.bar()`            |
| `plt.hist()`    | `ax.hist()`           |
| `plt.pie()`     | `ax.pie()`            |
| `plt.title()`   | `ax.set_title()`      |
| `plt.xlabel()`  | `ax.set_xlabel()`     |
| `plt.ylabel()`  | `ax.set_ylabel()`     |
| `plt.grid()`    | `ax.grid()`           |
| `plt.legend()`  | `ax.legend()`         |

## 2. Histogram
show the distribution

| Attribute     | Purpose                      | Example             |
| ------------- | ---------------------------- | ------------------- |
| `bins`        | Number or edges of intervals | `bins=10`           |
| `color`       | Fill color                   | `color="red"`       |
| `edgecolor`   | Border color                 | `edgecolor="black"` |
| `linewidth`   | Border thickness             | `linewidth=2`       |
| `alpha`       | Transparency (0–1)           | `alpha=0.6`         |
| `histtype`    | Histogram style              | `"bar"`, `"step"`   |
| `orientation` | Vertical/Horizontal          | `"horizontal"`      |
| `density`     | Show probability density     | `True`              |
| `cumulative`  | Running total                | `True`              |
| `label`       | Legend label                 | `"Marks"`           |
| `rwidth`      | Relative bar width           | `0.8`               |
| `align`       | Bin alignment                | `"mid"`             |
| `log`         | Logarithmic frequency axis   | `True`              |
| `zorder`      | Drawing order                | `zorder=3`          |


| `align` Value | Meaning                                |
| ------------- | -------------------------------------- |
| `"mid"`       | Bar is centered in the bin (default)   |
| `"left"`      | Bar starts at the left edge of the bin |
| `"right"`     | Bar ends at the right edge of the bin  |


```python
marks = [45,47,49,52,54,55,58,61,62,65,
         68,71,72,75,78,82,85,88,90]
# bins means range of nos like 50-60
class_A = [45,47,49,52,54,55,58,61,62,65,68,71,72,75,78]
plt.hist(marks,color="g",zorder=10,bins=10,edgecolor="w",linewidth="2",rwidth=1,histtype="bar",orientation="vertical",label="marks",align="left",
         # log=True,
         # density=True,
         # cumulative=True,
         alpha=0.7)
plt.hist(class_A, color="blue", alpha=0.5, label="Class A",edgecolor="k")
plt.legend()
plt.xlabel("marks",fontsize=20)
plt.ylabel("no. of strudents",fontsize=20)
plt.title("Histogram",fontsize=20,fontweight="bold",color="c")
plt.grid(zorder=25)
# By default (or depending on the style you're using, like ggplot), grid lines are often drawn below plot artists, regardless of the grid's zorder.
# align decides where each histogram bar is positioned relative to its bin.
plt.show()
```

```python
class_A = [45,47,49,52,54,55,58,61,62,65,68,71,72,75,78]
class_B = [50,52,55,57,60,63,65,68,70,73,76,79,82,85,88]

plt.hist([class_A, class_B],
         bins=10,
         color=["blue","red"],
         alpha=0.8,
         label=["Class A","Class B"])

plt.legend()
plt.show()
```

# 3. Box plot
summarizes the data using just five important numbers
| Term                              | Meaning                                                              | Easy to Remember     |
| --------------------------------- | -------------------------------------------------------------------- | -------------------- |
| **Minimum**                       | Smallest value that is **not an outlier**                            | Start of the whisker |
| **Q1 (25th Percentile)**          | Value below which **about 25% of the observations** lie              | Bottom of the box    |
| **Median (Q2 / 50th Percentile)** | Middle value of the data; about **50% of observations are below it** | Line inside the box  |
| **Q3 (75th Percentile)**          | Value below which **about 75% of the observations** lie              | Top of the box       |
| **Maximum**                       | Largest value that is **not an outlier**                             | End of the whisker   |


| Attribute                                    | Options                          | Description                                                         |
| -------------------------------------------- | -------------------------------- | ------------------------------------------------------------------- |
| `vert`                                       | `True`, `False`                  | Vertical (`True`) or horizontal (`False`) box plot *(older API)*    |
| `orientation`                                | `"vertical"`, `"horizontal"`     | Preferred in newer Matplotlib versions instead of `vert`            |
| `patch_artist`                               | `True`, `False`                  | Fill the box with color (`True`) or draw only the outline (`False`) |
| `showmeans`                                  | `True`, `False`                  | Display the **mean** as a marker                                    |
| `showfliers`                                 | `True`, `False`                  | Show or hide outliers                                               |
| `notch`                                      | `True`, `False`                  | Draw a notch around the median (dress like shape)                                     |
| `widths`                                     | Float (e.g. `0.3`, `0.6`, `0.8`) | Controls the width of the box (default=0.2)                                     |
| `labels` *(older)* / `tick_labels` *(newer)* | List of strings                  | Labels for multiple box plots                                       |
| `whis`                                       | Float, tuple, `"range"`          | Controls whisker length (default `1.5`)                             |
| `boxprops`                                   | Dictionary                       | Customize the box (color, linewidth, linestyle, etc.)               |
| `medianprops`                                | Dictionary                       | Customize the median line                                           |
| `whiskerprops`                               | Dictionary                       | Customize the whiskers                                              |
| `capprops`                                   | Dictionary                       | Customize the whisker caps                                          |
| `flierprops`                                 | Dictionary                       | Customize outlier markers                                           |
| `meanprops`                                  | Dictionary                       | Customize the mean marker (used with `showmeans=True`)              |
| `manage_ticks`                               | `True`, `False`                  | Automatically manage axis ticks (rarely changed)                    |
| `zorder`                                     | Integer                          | Drawing order of the box plot                                       |
<!--  -->
label is for legend and tick_labels=[]is for x-axis

```python
marks = [45,47,49,52,54,55,58,61,62,65,
         68,71,72,75,78,82,85,88,90,125]
plt.style.use("default") 
plt.boxplot(marks,notch=False,vert=True,widths=0.3,patch_artist=False,tick_labels=["marks"],showmeans=True,whis=1.5, boxprops=dict(color='blue', linewidth=2),
    whiskerprops=dict(color='red', linestyle='--'),
    capprops=dict(color='green', linewidth=2),
    medianprops=dict(color='black', linewidth=3),
    flierprops=dict(marker='o', markerfacecolor='purple'))
# plt.legend()
plt.grid(linestyle=":")
plt.show()
```

## multiple boxplots

```python
class_A = [55, 60, 62, 65, 67, 70, 72, 75, 78, 80]
class_B = [45, 50, 55, 58, 60, 62, 65, 68, 70, 72]
data=[class_A,class_B]
plt.boxplot(data,notch=False,vert=True,widths=0.3,patch_artist=False,tick_labels=["class_A","class_B"],showmeans=True,whis=1.5, boxprops=dict(color='blue', linewidth=2),
    whiskerprops=dict(color='red', linestyle='--'),
    capprops=dict(color='green', linewidth=2),
    medianprops=dict(color='black', linewidth=3),
    flierprops=dict(marker='o', markerfacecolor='purple'))
# plt.legend()
plt.grid(linestyle=":")
plt.show()
```

# 4. Pie chart
to show how each category contributes to the whole (100%) as slices of a circle

| Parameter       | Purpose           | Example                 |
| --------------- | ----------------- | ----------------------- |

## Multiple bar chart

Matplotlib can only place one bar per category automatically. When you want multiple bars for the same category, you need numeric positions (0, 1, 2, ...) so you can shift each group's bars slightly left or right.

```python
subjects = ["Math", "Science", "English", "Hindi"]

boys = [80, 75, 90, 70]
girls = [85, 90, 88, 80]

x = np.arange(len(subjects))
print(x)
width = 0.4

plt.bar(x, boys, width=0.4, label="Boys", color="skyblue")
plt.bar(x+width, girls, width=0.4, label="Girls", color="pink")

plt.xticks(x, subjects)
# this again converts numbers into subject names

plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Marks Comparison")

plt.legend()

plt.show()
```

# 6.Scatter plot

to show the relationship between two numerical variables.

| **Parameter** | **Accepted Values**                                                                | **Short Explanation**                                |
| ------------- | ---------------------------------------------------------------------------------- | ---------------------------------------------------- |
| `x`           | List, NumPy array, Pandas Series                                                   | Values for the X-axis (Required).                    |
| `y`           | List, NumPy array, Pandas Series                                                   | Values for the Y-axis (Required).                    |
| `s`           | Number (`100`) or list (`[20,40,60]`)                                              | Controls the size of the markers.                    |
| `c`           | Color name (`"red"`), hex (`"#FF0000"`), RGB tuple, list of colors, numeric values | Sets the color of the markers.                       |
| `marker`      | `"o"`, `"*"`, `"^"`, `"s"`, `"D"`, `"+"`, `"x"`, `"p"`, `"h"`                      | Changes the marker shape.                            |
| `alpha`       | Float (`0` to `1`)                                                                 | Controls marker transparency.                        |
| `edgecolors`  | Color name, `"none"`                                                               | Sets the border color of markers.                    |
| `linewidths`  | Number (`1`, `2`, `3`)                                                             | Sets the border thickness of markers.                |
| `label`       | String                                                                             | Label used for the legend.                           |
| `cmap`        | `"viridis"`, `"plasma"`, `"cool"`, `"hot"`, `"rainbow"`, `"winter"`                | Applies a colormap when `c` contains numeric values. |
| `vmin`        | Number                                                                             | Minimum value for color mapping.                     |
| `vmax`        | Number                                                                             | Maximum value for color mapping.                     |
| `zorder`      | Integer                                                                            | Controls drawing order (higher appears on top).      |


```python
hours = [1, 2, 3, 4, 5, 6]
marks = [35, 80, 50, 65, 75, 120]
plt.scatter(hours,marks,s=120,alpha=0.8,edgecolors="k",linewidths=2,marker="^",c=hours,cmap="rainbow")
plt.colorbar()
# give info about colors
plt.xlabel("hours")
plt.ylabel("marks")
plt.title("scatter plot",fontweight="bold",fontsize=15)
plt.show()
```

# 8. Subplots
A subplot in Matplotlib allows you to display multiple plots in a single figure. It is useful when you want to compare different graphs side by side or one above another.
#### plt.subplot(nrows, ncols, index)

```python
x = [1,2,3,4]
y = [10,20,15,25]

plt.subplot(2,2,1)
plt.grid(linestyle=":")
plt.scatter(x,y,c="r")
plt.title("scatter plot",fontweight="bold",color="green")

plt.subplot(2,2,2)
plt.bar(x,y,color="y")
plt.xticks(x,["jan","Feb","march","april"])
plt.title("bar plot",fontweight="bold",color="blue")

plt.subplot(2,2,3)
plt.pie([67,23,98,50,32,21],labels=["A","B","C","D","E","F"],radius=1.2 )
plt.title("Pie Chart",fontweight="bold",color="k")


plt.subplot(2,2,4)
plt.boxplot([x,y],tick_labels=["class_A","class_B"],patch_artist=True )
plt.title("box plot",fontweight="bold",color="c")

plt.tight_layout()
# to correct overlapping
plt.show()
```

# 9. SaveFig

| Parameter     | Description                      | Example               |

# 10. object oriented Style
The Object-Oriented style creates plots by using Figure and Axes objects instead of relying on the global pyplot state.
<!--  -->
fig and ax are objects.Objects have their own methods.

| Method            | Purpose                  |
| ----------------- | ------------------------ |
| `ax.plot()`       | Line plot                |
| `ax.scatter()`    | Scatter plot             |
| `ax.bar()`        | Bar chart                |
| `ax.hist()`       | Histogram                |
| `ax.pie()`        | Pie chart                |
| `ax.boxplot()`    | Box plot                 |
| `ax.imshow()`     | Display image or heatmap |
| `ax.set_title()`  | Set title                |
| `ax.set_xlabel()` | X-axis label             |
| `ax.set_ylabel()` | Y-axis label             |
| `ax.legend()`     | Show legend              |
| `ax.grid()`       | Show grid                |
<!--  -->
| Method                  | Purpose                        |
| ----------------------- | ------------------------------ |
| `fig.savefig()`         | Save the figure                |
| `fig.suptitle()`        | Overall title for all subplots |
| `fig.set_size_inches()` | Change figure size             |


#### The figure controls the entire canvas.
It is responsible for
<!--  -->
Figure size
<!--  -->
Background color
<!--  -->
DPI
<!--  -->
Saving image
<!--  -->
Overall title
<!--  -->
##### fig, ax = plt.subplots(rows, columns, figsize=(width, height))

```python
x = [1,2,3,4]
y = [10,20,15,25]
fig,ax=plt.subplots(1,2,figsize=(10,4), dpi=150)
# This single line creates:
# fig → the whole page(width,height)
# ax → one graph area on that page
ax[0].plot(x, y,marker="o",label="sales",color="g")
ax[0].legend()
ax[0].grid(linestyle=":")
ax[0].set_title("Line graph",color="r")
ax[0].set_xlabel("months")
ax[0].set_ylabel("sale")
ax[0].set_ylim(5, 26)
# to display number on y axis in between this range

ax[1].bar(x, y,label="sales",color="c")
ax[1].legend()
ax[1].grid(linestyle=":")
ax[1].set_title("Bar graph",color="m")
ax[1].set_xlabel("months")
ax[1].set_ylabel("sale")
ax[1].set_xticks(x,["Jan","Feb","March","April"])

fig.suptitle("Matplotlib Object-Oriented Example", fontsize=16,color="orange",fontweight="bold")
fig.tight_layout()

plt.show()
```

| `plt.subplots()`            | `fig.add_axes()`                           |
| --------------------------- | ------------------------------------------ |
| Automatic grid layout       | Manual positioning                         |
| Easy to use                 | More control                               |
| Most commonly used          | Used for custom layouts                    |
| Dashboards, multiple graphs | Insets, zoomed graphs, custom arrangements |
<!--  -->
#### axes=fig.add_axes([left, bottom, width, height])

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8,5))

ax1 = fig.add_axes([0.1,0.1,0.35,0.8])

ax2 = fig.add_axes([0.6,0.1,0.3,0.3])

ax1.plot([1,2,3],[3,2,5])

ax2.bar([1,2,3],[4,6,2])

plt.show()
```
